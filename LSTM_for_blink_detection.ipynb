## Final data simulation for LSTM:

### on data

```python
import numpy as np
on_data=np.ones((150,150,1))
print(on_data.shape)

# off data
off_data=np.zeros((150,150,1))
off_data.shape
```

```python

Label_on=np.ones(150)
Label_on=np.where(Label_on==1, 'on', Label_on)
print(Label_on.shape)

Label_off=np.zeros(150)
Label_off=np.where(Label_off==0, 'off', Label_off)
Label_off.shape

```

### slow blink data

```python
# create 15 zeros and 15 ones 5 times (total=150):
from collections import deque
Ones_slow=np.ones(15)
Zeros_slow=np.zeros(15)

Final_slow=np.block([Ones_slow,Zeros_slow]*5)

# array anti clockwise rotate:
list_slow=[]
for i in range(len(Final_slow)):
  # converting list into deque to rotate
  Final_slow=deque(Final_slow) 
  Final_slow.rotate(-1)
  #converting deque into list of array
  Final_slow=list(Final_slow)
  list_slow.append(Final_slow)
list_slow

```

```python
#converting list into array:
array_slow=np.array(list_slow)
```

```python
array_slow.shape
```

```python
# create labels in array:
Label_slow=np.zeros(150)
Label_slow=np.where(Label_slow==0, 'slow_blink', Label_slow)
Label_slow=np.array(Label_slow)
Label_slow
```

```python
Label_slow.shape
```

### fast blink data

```python
# create 15 zeros and 15 ones 5 times (total=150):
Ones_fast=np.ones(8)
Zeros_fast=np.zeros(8)
e_fast=np.ones(6)

pattern=np.block([Ones_fast,Zeros_fast]*9)

Final_fast=np.block([pattern,e_fast])

# array anti clockwise rotate:
list_fast=[]
for i in range(len(Final_fast)):
  # converting list into deque to rotate
  Final_fast=deque(Final_fast) 
  Final_fast.rotate(-1)
  #converting deque into list of array
  Final_fast=list(Final_fast)
  list_fast.append(Final_fast)
list_fast
```

```python
#converting list into array:
array_fast=np.array(list_fast)
array_fast.shape
```

```python
# create labels in array:
Label_fast=np.zeros(150)
Label_fast=np.where(Label_fast==0, 'fast_blink', Label_fast)
Label_fast=np.array(Label_fast)
Label_fast
```

```python
Label_fast.shape
```

# Preprocessing data for LSTM

Now we have created array_slow,label_slow, array_fast, label_fast

```python
array_slow=array_slow.reshape(150,150,1)
array_fast=array_fast.reshape(150,150,1)
```

```python
print(array_slow.shape)
print(array_fast.shape)
print(on_data.shape)
off_data.shape
```

```python
X = np.concatenate([array_slow, array_fast,on_data,off_data],axis=0)
X.shape
```

```python
y = np.concatenate([Label_slow, Label_fast,Label_on,Label_off],axis=0)
y.shape
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

label_En = LabelEncoder()
label = label_En.fit_transform(y)
print(label.shape)
label
```

```python
#  fast_blink=0, off=1, on=2, slow_blink=3
label_En.classes_
```

```python
from tensorflow.keras.utils import to_categorical

y = to_categorical(label).astype(int)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

# Build and Train LSTM Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import TensorBoard
```

```python
import os

log_dir = os.path.join('Logs1')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
model = Sequential()
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(150,1)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(4, activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20, callbacks=[tb_callback])
```

```python
model.summary()
```

# Make predictions

```python
y_pred = model.predict(X_test)

```

# Save weights

```python
model.save('led_state.h5')
```
